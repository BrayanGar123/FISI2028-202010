```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv('Cars93.csv')
```

```python
plt.figure(figsize=(25,4))

plt.subplot(1,4,1)
plt.scatter(data['Horsepower'], data['Price'] )
plt.ylabel('Price')
plt.xlabel('Horsepower')

plt.subplot(1,4,2)
plt.scatter(data['Length'], data['Price'])
plt.ylabel('Price')
plt.xlabel('Length')

plt.subplot(1,4,3)
plt.scatter(data['Fuel.tank.capacity'], data['Price'])
plt.ylabel('Price')
plt.xlabel('Fuel.tank.capacity')

plt.subplot(1,4,4)
plt.scatter( data['Width'], data['Price'])
plt.xlabel('Width')
plt.ylabel('Price')
```
