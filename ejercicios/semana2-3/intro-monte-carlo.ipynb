# Ejercicio

Hacer una función que calcule el valor de $\pi$ en función del número de puntos N

Graficar el error entre el valor esperado y el valor obtenido

```python
def ValorDePi(N):
    N = int(N)
    x_r = np.random.rand(N)
    y_r = np.random.rand(N)
    
    p_bajo_c = 0
    for i in range(N):
        if( (x_r[i] **2 + y_r[i]**2) < 1):
            p_bajo_c += 1
    return( 4 * p_bajo_c / N )
```

```python
ValorDePi(1000)
```

```python
N = np.linspace(1,1000,100)

err = []

for n in N:
    err.append( abs(np.pi - ValorDePi(n)))
```

```python
plt.plot(N, err)
plt.xscale("log")
```

```python
N = np.logspace(1,6)
err = []
for n in N:
    err.append( abs(np.pi - ValorDePi(n)))
```

```python
plt.plot(N, err)
plt.xscale("log")
plt.yscale("log")
```
